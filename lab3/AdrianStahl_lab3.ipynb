##### Biorę plik results.csv z laboratoriów 1, dot. iloczynu skalarnego macierzy oraz mnozenia macierzy.

```r
# library("tidyverse") - ponizej są biblioteki które zawierają się właśnie w 'tidyverse'
library(dplyr)
library(tidyr)
library(ggplot2)
```

```r
data <- read.csv("results.csv")
```

```r
data
```

```r
vectorData <- data %>% group_by(type, size) %>% summarize(mean=mean(time), std=sd(time), .groups="drop") %>% filter (type=="iloczyn_skalarny")
matrixData <- data %>% group_by(type, size) %>% summarize(mean=mean(time), std=sd(time), .groups="drop") %>% filter (type=="mnozenie_macierzy")
```

```r
vectorData
```

```r
matrixData
```

```r
plotVector <- ggplot(vectorData, aes(x = size, y = mean, ymin = mean - std, ymax = mean + std)) +
  geom_point() +
  geom_errorbar(width = 0.2) +
  labs(x = "dl. wektora", y = "pomiar (s)", title = "Iloczyn skalarny")
plotVector
```

```r
plotMatrix <- ggplot(matrixData, aes(x = size, y = mean, ymin = mean - std, ymax = mean + std)) +
  geom_point() +
  geom_errorbar(width = 0.2) +
  labs(x = "rozmiar macierzy", y = "pomiar (s)", title = "Mnozenie macierzy")
plotMatrix
```

```r
library(gridExtra)

# Arrange the plots side-by-side
grid.arrange(plotVector, plotMatrix, ncol = 2)
```
